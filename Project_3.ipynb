This section imports the make_classification method and creates the input dataset to be used by the neural networks.

```python
from sklearn.datasets import make_classification
X, y = make_classification(n_samples=200, n_features=12)

print("Input array: ", X)
print("Size of input array: ", len(X))
print("Number of input nodes: ", len(X[0]))
```

This section imports the necessary tensorflow packages and creates the full neural network.

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

# Full neural network
full_model = keras.models.Sequential([layers.Dense(8, activation="sigmoid", input_shape=(12,)),
                                      layers.Dense(4, activation="sigmoid"),
                                      layers.Dense(8, activation="sigmoid"),
                                      layers.Dense(12, activation="sigmoid")])
full_output = full_model(X)
print(full_output)
```

This section creates the first-half neural network.

```python
# First-half neural network
first_half = keras.models.Sequential([layers.Dense(8, activation="sigmoid", input_shape=(12,)),
                                      layers.Dense(4, activation="sigmoid")])
fhalf_output = first_half(X)
print(fhalf_output)
```

This section creates the second-half neural network.

```python
# Second-half neural network
second_half = keras.models.Sequential([layers.Dense(8, activation="sigmoid", input_shape=(4,)),
                                       layers.Dense(12, activation="sigmoid")])
shalf_output = second_half(fhalf_output)
print(shalf_output)
```

This section calculates the error for each node of the full neural network and then calculates the full error of the nodes for the full neural network.

```python
# Check error for full neural network
full_error = [None] * len(X)
total_full_error = 0;
for i in range(0, len(X)):
  error = 0
  for j in range(0, len(X[0])):
    error += 0.5*pow(X[i][j] - full_output[i][j], 2)
  full_error[i] = error
  total_full_error += error
  print("Error of node ", i, ": ", full_error[i])
print("Total full error of nodes: ", total_full_error)
```

This section calculates the error for each node of the 2 halves of the neural network and then calculates the total halves error of the nodes.

```python
# Check error for two halves of the neural network
halves_error = [None] * len(X)
total_halves_error = 0
for i in range(0, len(X)):
  error = 0
  for j in range(0, len(X[0])):
    error += 0.5*pow(X[i][j] - shalf_output[i][j], 2)
  halves_error[i] = error
  total_halves_error += error
  print("Error of node ", i, ": ", halves_error[i])
print("Total halves error of nodes: ", total_halves_error)
```
